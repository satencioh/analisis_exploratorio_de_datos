```python
import pandas as pd
import altair as alt
```

```python
data_chicago = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'
df = pd.read_csv(data_chicago)
```

```python
df['CRASH_DATE'] = df['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce',utc=True))
```

```python
df.info()
```

```python
report_1 = df.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'NUM_UNITS':'sum'})
```

```python
report_1 = report_1.reset_index()
```

```python
report_1
```

```python
alt.Chart(report_1).mark_bar().encode(x ='LIGHTING_CONDITION',y='NUM_UNITS', color='REPORT_TYPE').properties(width=200)
```

```python
report_2 = df.groupby(['LIGHTING_CONDITION','REPORT_TYPE','CRASH_HOUR']).agg({'NUM_UNITS':['sum','min','max']})
```
